```python
import csv
```

```python
hours_studied = []
previous_scores = []
sleep_hours = []
sample_papers = []
performance_index = []

with open("/content/StudentPerformance.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        hours_studied.append(float(row["Hours Studied"]))
        previous_scores.append(float(row["Previous Scores"]))
        sleep_hours.append(float(row["Sleep Hours"]))
        sample_papers.append(float(row["Sample Question Papers Practiced"]))
        performance_index.append(float(row["Performance Index"]))
```

```python
def mean(values):
    return sum(values) / len(values)

def variance(values):
    m = mean(values)
    return sum((x - m) ** 2 for x in values) / len(values)
```

```python
print("Mean Hours Studied:", mean(hours_studied))
print("Mean Previous Scores:", mean(previous_scores))
print("Mean Sleep Hours:", mean(sleep_hours))
print("Mean Sample Papers Practiced:", mean(sample_papers))

print("Mean Performance Index:", mean(performance_index))
print("Variance of Performance Index:", variance(performance_index))
```
